e considere as fórmulas gerais da regressão linear simples



$$\hat{w}_0 = 1/N\sum_i^N{y_i}- 1/N\hat{w}_1\sum_i^N{x_i}$$


$$\hat{w}_1 = \frac{N\sum_i^N{y_ix_i} - \sum_i^N{x_i}\sum_i^N{y_i}}{ N\sum_i^N{x^2_i} - \sum_i^N{x_i}\sum_i^N{x_i}}$$



a) Determine qual a estimativa de idade para um modelo de regressão linear para uma instância com 25 $\mu$m de comprimento.

b) Caso existisse mais uma *feature* neste problema, como estenderia o modelo à segunda feature?

````python
%matplotlib notebook
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


x = np.array([69.84, 37.24, 44.75, 94.78, 9.43]).reshape(-1,1)
y = np.array([374.32, 253.87, 253.42, 509.59, 74.22]).reshape(-1,1)


R = np.hstack([np.ones((x.shape)).reshape(-1,1),x])

#w_fit = (R^TR)^{-1}R^Ty
y_intermediate = np.matmul(R.T,y)
w_fit = np.matmul(np.linalg.inv(np.matmul(R.T,R)),y_intermediate)

y_fit = w_fit[0]+w_fit[1]*x

fig = plt.figure()
plt.plot(x,y,".")
plt.plot(x,y_fit,"-")
plt.show()

w_fit
````

````python
w1 = (5*sum(x*y) - sum(x)*sum(y))/(5*sum(x*x)-sum(x)*sum(x))

w0 = sum(y)/5 - sum(x)*w1/5

w1, w0
````
